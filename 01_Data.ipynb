## Data  
**Data** = The collected observations we have about something
- **Continuous**: stock price
- **Categorical**: What car has the best repair history  

<br></br>
### Types of Measurement
**Nominal**
- Predetermined categories
- Can not be sorted  
    - E.g Animal classification
    - Political party

**Ordinal**
- Can be sorted
- Lacks scale
- A kin to ordered categories
   - Survery responces (Often, Sometimes, Selson, Never)

**Iterval**
- Provides a scale
- No "Zero" point, i.e you can go negative
    - Temperature (Celsius or Fahrenheit)
    
**Ratio**
- Values that have a true "Zero" point, i.e you can not go negative
    - Age
    - Weight
    - Salary

<br></br>
## Central Tendency
**Mean**: calculated average
- Can not describe the shape of the total data
- Can be influenced by outliers

**Median**: Middle value 
- Can provide a closer value to the other values in the data set if an outlier is present

**Mode**: Most frequently occuring value

<br></br>
## Dispersion
Describes the spread of the data

**Range**: 
- Maximum value - minimum value

**Variance**: 
- Sum of the squared distance from each point to the mean, divided by the population
- Difference between SAMPLE variance and the POPULATION variance
    - Subject to Bessel's correction (n-1)

**Standard deviation**: 
- The sqaure root of the variance
    - Variance: s^2
    - Standard deviation: s

## Quartiles and the Inter Quartile Range (IQR)
- Every data point is considered, not aggregated
- Box plots can be created using quartiles
- Fences and outliers
    - **IQR**: Q3 - Q2 i.e. the range of the box in the box-plot
    - **Fence** : 1.5 x IQR either side of Q1 and Q3
    - Any value outside the fence is an outlier
