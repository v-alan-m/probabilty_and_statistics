## Probabilty
- A value between 0 and 1 that a certain event will occur
    - Probabilty can be multipled by 100 to give a percentage
        - E.g P(COIN_FLIP<sub>heads</sub>) = 0.5 = 50%
- When an action is taken to calculate the probabilty it is called a **trail**
    - Trails are independent events
- Each trail of flippig a coin: **experiment**
- Each mutually exclusive outcomeL **simple event**
- The sum total of every possible simple events: **sample space**

<br></br>
### Permutations
- An arangment of as set of objects in a certain order
- The number possible different permutations of rearranging 'ABC' can be caluclated using *n!*
    - 'abc' = 3 items = 3!
    - 3! = 3 x 2 x 1 = 6 permutations 
        - abc, acb, bac, bca, cab, cba

- Take a subset of the items in a permutation
    - n: total number of objects in the set
    - r: the size of the sub-set

### Bayes Theorm
- Ued to determine the *probabilty* of a **parameter given a certain event**  
<img src='pics/bayes_theorem.png'>  
<br></br>
Medical field Example
- 1 out of 500 products are defective (0.2%)
- Tools identifies a defective part 99% of the time
- What is the probabilty that the part really is defective?  
<br></br>
- Method:
    - P(A|B): Defective part receiving postive test-result (True positive)
    - P(B|A): Defective part receiving negative test-result (False positive)
    - P(A): Defective part 
    - P(B): Positive test-result  
<br></br>
- Calculation:
    - P(A|B): ?
    - P(B|A): 0.99
    - P(A): 0.002
    - P(B): ?  

<br></br>
*Remember:*
- **Multiply** *for probabilty given another probabilty*
- **Add** *for probabilty follwed by another probabilty*

<img src='pics/medical_prob_b.png'>

- Expanded version of Bayes Theorem  

<img src='pics/bayes_theorem_exp.png'>

- P(A|B) = (0.99*0.002)/(0.99*0.002+0.02*0.998) = 0.165
    - P(A|B) = 16.5%
    - There is a 16.5% chance that a defective part will be identified and that the part really is defective  
<br></br>
- If the test is performed a 2nd time
    - P(A) = 0.165

<img src='pics/medical_second_test.png'>

- **P(A|B) = 0.951 = 95.1%**
